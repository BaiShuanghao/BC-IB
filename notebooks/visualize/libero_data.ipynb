```python
import h5py
import cv2
import numpy as np
import os
```

```python
file_path = '/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_goal/open_the_middle_drawer_of_the_cabinet_demo.hdf5'
with h5py.File(file_path, 'r') as f:
    def explore_hdf5(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(f"数据集: {name}, 形状: {obj.shape}, 数据类型: {obj.dtype}")
        elif isinstance(obj, h5py.Group):
            print(f"组: {name}")

    f.visititems(explore_hdf5)
```

```python
# 文件路径
file_path = "/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_goal/open_the_middle_drawer_of_the_cabinet_demo.hdf5"
output_path = 'video.mp4'

# 打开 HDF5 文件
with h5py.File(file_path, 'r') as f:
    # 访问数据集
    dataset_name = "data/demo_0/obs/agentview_rgb"
    images = f[dataset_name][:] 
    
    # 确保图片格式为 BGR（OpenCV 使用 BGR 格式）
    images_bgr = [cv2.cvtColor(img, cv2.COLOR_RGB2BGR) for img in images]
# 定义视频编码器和参数
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 使用 MP4 编码器
fps = 10  # 帧率，每秒 10 帧
frame_size = (images[0].shape[1], images[0].shape[0])  # 图片的宽度和高度

# 初始化视频写入器
video_writer = cv2.VideoWriter(output_path, fourcc, fps, frame_size)

# 写入每帧到视频文件
for frame in images_bgr:
    video_writer.write(frame)

# 释放视频写入器
video_writer.release()

print(f"视频已成功保存到 {output_path}")
```

```python
# 文件路径
file_path = "/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_goal/open_the_middle_drawer_of_the_cabinet_demo.hdf5"
file_path = "/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_object/pick_up_the_butter_and_place_it_in_the_basket_demo.hdf5"
file_path = "/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_spatial/pick_up_the_black_bowl_on_the_stove_and_place_it_on_the_plate_demo.hdf5"
file_path = "/baishuanghao/code/BC-IB/third_party/benchmarks/LIBERO/libero/datasets/libero_10/KITCHEN_SCENE3_turn_on_the_stove_and_put_the_moka_pot_on_it_demo.hdf5"

# 输出图片保存路径
output_folder = "output_images"
os.makedirs(output_folder, exist_ok=True)  # 确保输出文件夹存在

# 打开 HDF5 文件
with h5py.File(file_path, 'r') as f:
    # 访问数据集
    dataset_name = "data/demo_0/obs/agentview_rgb"
    images = f[dataset_name][:]  # 读取所有图片，形状: (138, 128, 128, 3)
    
    # 每隔 20 帧保存一张图片
    for i in range(0, len(images), 20):  # 每隔 20 帧
        img = images[i]  # 获取第 i 帧
        img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  # 转换为 BGR 格式
        img_bgr = cv2.flip(img_bgr, 0)

        # 调整图片大小为 (224, 224)
        resized_img = cv2.resize(img_bgr, (224, 224), interpolation=cv2.INTER_LINEAR)
        
        # 保存图片到文件夹
        output_path = os.path.join(output_folder, f"frame_{i:03d}.png")
        cv2.imwrite(output_path, resized_img)
        print(f"Saved: {output_path}")

    # 处理最后一帧
    last_frame = images[-1]  # 获取最后一帧
    last_frame_bgr = cv2.cvtColor(last_frame, cv2.COLOR_RGB2BGR)  # 转换为 BGR 格式
    last_frame_bgr = cv2.flip(last_frame_bgr, 0)  # 翻转图片（上下翻转）
    resized_last_frame = cv2.resize(last_frame_bgr, (224, 224), interpolation=cv2.INTER_LINEAR)

    # 保存最后一帧图片
    last_output_path = os.path.join(output_folder, "frame_last.png")
    cv2.imwrite(last_output_path, resized_last_frame)
    print(f"Saved last frame: {last_output_path}")
```
